```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Sample data
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)

ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    line.set_data(x[:frame], y[:frame])
    return line,

ani = FuncAnimation(fig, update, frames=len(x), init_func=init, interval=20)

ani.save("animation.gif", writer="pillow")
plt.show()
```

## Using Plotly (interactive animations)

```python
import plotly.express as px

df = px.data.gapminder()
fig = px.scatter(
    df, x="gdpPercap", y="lifeExp",
    animation_frame="year", animation_group="country",
    size="pop", color="continent", log_x=True
)
fig.show()
```

## Using Manim (professional-quality animations)

```python
from manim import *

class CircleMove(Scene):
    def construct(self):
        dot = Dot()
        self.play(dot.animate.shift(RIGHT * 3))
        self.play(dot.animate.shift(UP * 2))
```
