## A* Solver Notebook

### Cell 1: Load Maze


```python
def load_maze(filename="maze.txt"):
    with open(filename, "r") as f:
        maze = [list(line.rstrip("\n")) for line in f]
    return maze
```

### Cell 2: Find Start & End

```python
def find_point(maze, symbol):
    for y, row in enumerate(maze):
        for x, cell in enumerate(row):
            if cell == symbol:
                return (x, y)
    return None
```

### Cell 3: Heuristic Function (Manhattan Distance)

```python
def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
```

### Cell 4: A* Algorithm

```python
import heapq

def astar(maze):
    start = find_point(maze, "S")
    end = find_point(maze, "E")

    open_set = []
    heapq.heappush(open_set, (0, start))

    came_from = {}
    g_score = {start: 0}

    while open_set:
        _, current = heapq.heappop(open_set)

        if current == end:
            return came_from, end

        x, y = current
        for dx, dy in [(0,1), (0,-1), (1,0), (-1,0)]:
            nx, ny = x + dx, y + dy
            neighbor = (nx, ny)

            if maze[ny][nx] == "#":
                continue

            tentative_g = g_score[current] + 1

            if tentative_g < g_score.get(neighbor, float("inf")):
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g
                f_score = tentative_g + heuristic(neighbor, end)
                heapq.heappush(open_set, (f_score, neighbor))

    return None, None
```

### Cell 5: Mark A* Path

```python
def mark_path(maze, came_from, end):
    current = end
    while current in came_from:
        x, y = current
        if maze[y][x] not in ("S", "E"):
            maze[y][x] = "."
        current = came_from[current]
```

### Cell 6: Print Maze

```python
def print_maze(maze):
    for row in maze:
        print("".join(row))
```

### Cell 7: Run A*

```python
maze = load_maze()

print("Original Maze:")
print_maze(maze)

came_from, end = astar(maze)

if came_from:
    mark_path(maze, came_from, end)
    print("\nSolved Maze (A*):")
    print_maze(maze)
else:
    print("No path found.")
```
