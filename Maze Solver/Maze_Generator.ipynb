## Notebook 1: Maze Generator (DFS)

### Cell 1: Imports & Constants

```python
import random

WIDTH = 21   # must be odd
HEIGHT = 21  # must be odd

WALL = "#"
PATH = " "
START = "S"
END = "E"
```

### Cell 2: Create Empty Maze

```python
def create_maze(width, height):
    maze = [[WALL for _ in range(width)] for _ in range(height)]
    return maze
```

### Cell 3: Maze Generation (DFS / Recursive Backtracker)

```python
def generate_maze(maze, x, y):
    directions = [(0, 2), (0, -2), (2, 0), (-2, 0)]
    random.shuffle(directions)

    maze[y][x] = PATH

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 < nx < WIDTH-1 and 0 < ny < HEIGHT-1:
            if maze[ny][nx] == WALL:
                maze[y + dy//2][x + dx//2] = PATH
                generate_maze(maze, nx, ny)
```

### Cell 4: Add Start and End

```python
def add_start_end(maze):
    maze[1][1] = START
    maze[HEIGHT-2][WIDTH-2] = END
```

### Cell 5: Save Maze to Text File

```python
def save_maze(maze, filename="maze.txt"):
    with open(filename, "w") as f:
        for row in maze:
            f.write("".join(row) + "\n")
```

### Cell 6: Run Generator

```python
maze = create_maze(WIDTH, HEIGHT)
generate_maze(maze, 1, 1)
add_start_end(maze)
save_maze(maze)

print("Maze saved to maze.txt")
```
