# Notebook 2: Sudoku Solver (From File)

## Cell 1: Load Sudoku from File

```python
def load_from_file(filename="sudoku.txt"):
    grid = []
    with open(filename, "r") as f:
        for line in f:
            grid.append([int(num) for num in line.split()])
    return grid
```

## Cell 2: Solver Functions

```python
def find_empty(grid):
    for row in range(9):
        for col in range(9):
            if grid[row][col] == 0:
                return row, col
    return None

def is_valid(grid, row, col, num):
    if num in grid[row]:
        return False

    for i in range(9):
        if grid[i][col] == num:
            return False

    box_x = col // 3
    box_y = row // 3

    for i in range(box_y * 3, box_y * 3 + 3):
        for j in range(box_x * 3, box_x * 3 + 3):
            if grid[i][j] == num:
                return False

    return True
```

## Cell 3: Backtracking Solver

```python
def solve_sudoku(grid):
    find = find_empty(grid)
    if not find:
        return True

    row, col = find

    for num in range(1, 10):
        if is_valid(grid, row, col, num):
            grid[row][col] = num
            if solve_sudoku(grid):
                return True
            grid[row][col] = 0

    return False
```

## Cell 4: Pretty Print Sudoku

```python
def print_sudoku(grid):
    for i, row in enumerate(grid):
        if i % 3 == 0 and i != 0:
            print("-" * 21)
        for j, num in enumerate(row):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")
            print(num if num != 0 else ".", end=" ")
        print()
```

## Cell 5: Run Solver

```python
sudoku = load_from_file()
print("Original Puzzle:")
print_sudoku(sudoku)

if solve_sudoku(sudoku):
    print("\nSolved Puzzle:")
    print_sudoku(sudoku)
else:
    print("No solution exists.")
```
